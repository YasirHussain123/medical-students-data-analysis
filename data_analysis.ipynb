# Missing Values Fill
##### ✅ Numeric columns filled with mean/median
##### ✅ Categorical columns filled with mode (most frequent value)
##### This is a best practice since it handles each type correctly

```python
for i in df.select_dtypes(include = "object").columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
for i in df.select_dtypes(include = "float64").columns:
    df[i].fillna(df[i].mean(),inplace=True)
```

# 📦 Boxplot for Outlier Detections
##### ✅ Boxplot used for outlier detection
##### Example: IQR method or Z-score method


```python
sns.boxplot(x="BMI",data=df)
plt.show()
```

```python
q1=df["BMI"].quantile(0.25)
q3 = df["BMI"].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
min_range = q1-1.5*IQR
max_range = q3+1.5*IQR
min_range,max_range
```

```python
df = df[(df["BMI"] > q1-1.5*IQR) & (df["BMI"] <  q3+1.5*IQR)]
```
